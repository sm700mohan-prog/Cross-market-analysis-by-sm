```python
all_records = []
import requests
import time

base_url = "https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&sparkline=False&page="

for page_num in range(1, 6): # Loop for pages 1 to 5
    url = f"{base_url}{page_num}"
    response = requests.get(url)
    if response.status_code == 200:
        page_data = response.json()
        all_records.extend(page_data)
    else:
        print(f"Failed to fetch data for page {page_num}. Status code: {response.status_code}")
    time.sleep(10) # Increased delay to 10 seconds between requests to avoid rate limiting

cryptocurrency = all_records

print(f"Total records collected: {len(cryptocurrency)}")
print("First 3 records:")
display(cryptocurrency)
```

```python
import datetime

filtered_cryptocurrency = []

for coin in cryptocurrency:
    filtered_coin = {
        'id': coin.get('id'),
        'symbol': coin.get('symbol'),
        'name': coin.get('name'),
        'current_price': coin.get('current_price'),
        'market_cap': coin.get('market_cap'),
        'market_cap_rank': coin.get('market_cap_rank'),
        'total_volume': coin.get('total_volume'),
        'circulating_supply': coin.get('circulating_supply'),
        'total_supply': coin.get('total_supply'),
        'ath': coin.get('ath'),
        'atl': coin.get('atl')
    }

    # Extract only the date part from 'last_updated'
    last_updated_str = coin.get('last_updated')
    if last_updated_str:
        try:
            # Assuming format like '2026-02-06T13:19:11.648Z'
            dt_object = datetime.datetime.fromisoformat(last_updated_str.replace('Z', '+00:00'))
            filtered_coin['last_updated_date'] = dt_object.date().isoformat()
        except ValueError:
            filtered_coin['last_updated_date'] = None # Handle cases where date format might differ
    else:
        filtered_coin['last_updated_date'] = None

    filtered_cryptocurrency.append(filtered_coin)

cryptocurrency = filtered_cryptocurrency

print(f"Total filtered records: {len(cryptocurrency)}")
display(cryptocurrency)
```

```python
import pandas as pd

df_cryptocurrency = pd.DataFrame(cryptocurrency)

display(df_cryptocurrency)
```

```python
top_3_coin_ids = df_cryptocurrency.sort_values(by='market_cap_rank').head(3)['id'].tolist()
print(top_3_coin_ids)
```

```python
all_historical_data = []
import time
import requests

for coin_id in top_3_coin_ids:
    url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart?vs_currency=inr&days=365"
    response = requests.get(url)
    if response.status_code == 200:
        data = response.json()
        all_historical_data.append(data)
    else:
        print(f"Failed to fetch historical data for {coin_id}. Status code: {response.status_code}")
    time.sleep(10)

print(f"Collected historical data for {len(all_historical_data)} cryptocurrencies.")
```

```python
historical_prices_list = []

for i, coin_data in enumerate(all_historical_data):
    coin_id = top_3_coin_ids[i]

    # Check if 'prices' key exists in coin_data
    if 'prices' in coin_data:
        # Extract price entries from the current coin's historical data
        prices = coin_data['prices']

        for price_record in prices:
            timestamp = price_record[0]
            price_inr = price_record[1]

            # Convert timestamp (milliseconds) to datetime object and format to YYYY-MM-DD
            date = pd.to_datetime(timestamp, unit='ms').strftime('%Y-%m-%d')

            # Create a dictionary for the current record
            record = {
                'coin_id': coin_id,
                'date': date,
                'price_inr': price_inr
            }
            historical_prices_list.append(record)
    else:
        print(f"Warning: 'prices' data not found for coin_id: {coin_id}. Skipping this coin.")

print(f"Processed {len(historical_prices_list)} historical price records.")
```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('mydatabase.db')

query_combined = """
WITH crypto_prices_2025 AS (
    SELECT
        date,
        MAX(CASE WHEN coin_id = 'bitcoin' THEN price_inr ELSE NULL END) AS bitcoin_price,
        MAX(CASE WHEN coin_id = 'ethereum' THEN price_inr ELSE NULL END) AS ethereum_price,
        MAX(CASE WHEN coin_id = 'tether' THEN price_inr ELSE NULL END) AS tether_price
    FROM
        historical_prices
    WHERE
        STRFTIME('%Y', date) = '2025'
        AND coin_id IN ('bitcoin', 'ethereum', 'tether')
    GROUP BY
        date
),
stock_prices_2025 AS (
    SELECT
        date,
        MAX(CASE WHEN ticker = '^GSPC' THEN close ELSE NULL END) AS gspc_close,
        MAX(CASE WHEN ticker = '^IXIC' THEN close ELSE NULL END) AS ixic_close,
        MAX(CASE WHEN ticker = '^NSEI' THEN close ELSE NULL END) AS nsei_close
    FROM
        stock_prices
    WHERE
        STRFTIME('%Y', date) = '2025'
        AND ticker IN ('^GSPC', '^IXIC', '^NSEI')
    GROUP BY
        date
)
SELECT
    cp.date,
    cp.bitcoin_price,
    cp.ethereum_price,
    cp.tether_price,
    sp.gspc_close,
    sp.ixic_close,
    sp.nsei_close
FROM
    crypto_prices_2025 AS cp
LEFT JOIN
    stock_prices_2025 AS sp ON cp.date = sp.date
ORDER BY
    cp.date ASC;
"""

df_combined_crypto_stocks_2025 = pd.read_sql_query(query_combined, conn)

conn.close()

print("Combined daily prices for top 3 cryptocurrencies and stock indices for 2025:")
display(df_combined_crypto_stocks_2025.head())
print(f"DataFrame shape: {df_combined_crypto_stocks_2025.shape}")
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd

def app():
    st.set_page_config(layout="wide")
    st.title("Market Analysis Dashboard")

if __name__ == '__main__':
    app()
```

```python
%%writefile app.py
import streamlit as st
import pandas as pd
import sqlite3
import altair as alt
import datetime

# Database connection function
@st.cache_resource
def get_db_connection():
    conn = sqlite3.connect('mydatabase.db')
    return conn

# --- Data Loading Functions ---
@st.cache_data(ttl=600)
def load_cryptocurrency_data():
    conn = get_db_connection()
    df = pd.read_sql_query("SELECT * FROM cryptocurrency", conn)
    return df

@st.cache_data(ttl=600)
def load_historical_prices(coin_id=None, start_date=None, end_date=None):
    conn = get_db_connection()
    query = "SELECT coin_id, date, price_inr FROM historical_prices WHERE 1=1"
    params = []
    if coin_id:
        query += " AND coin_id = ?"
        params.append(coin_id)
    if start_date:
        query += " AND date >= ?"
        params.append(start_date)
    if end_date:
        query += " AND date <= ?"
        params.append(end_date)
    query += " ORDER BY date ASC"
    df = pd.read_sql_query(query, conn, params=params)
    df['date'] = pd.to_datetime(df['date'])
    return df

@st.cache_data(ttl=600)
def load_oil_prices(start_date=None, end_date=None):
    conn = get_db_connection()
    query = "SELECT date, price_usd FROM oil_prices WHERE 1=1"
    params = []
    if start_date:
        query += " AND date >= ?"
        params.append(start_date)
    if end_date:
        query += " AND date <= ?"
        params.append(end_date)
    query += " ORDER BY date ASC"
    df = pd.read_sql_query(query, conn, params=params)
    df['date'] = pd.to_datetime(df['date'])
    return df

@st.cache_data(ttl=600)
def load_stock_prices(ticker=None, start_date=None, end_date=None):
    conn = get_db_connection()
    query = "SELECT date, ticker, open, high, low, close, volume FROM stock_prices WHERE 1=1"
    params = []
    if ticker:
        query += " AND ticker = ?"
        params.append(ticker)
    if start_date:
        query += " AND date >= ?"
        params.append(start_date)
    if end_date:
        query += " AND date <= ?"
        params.append(end_date)
    query += " ORDER BY date ASC"
    df = pd.read_sql_query(query, conn, params=params)
    df['date'] = pd.to_datetime(df['date'])
    return df


# --- Page 1: Filters & Data Exploration ---
def filters_and_data_exploration_page():
    st.title("Filters & Data Exploration")

    st.header("Global Market Snapshot")

    # Date Range Selector
    col1, col2 = st.columns(2)
    with col1:
        default_start_date = datetime.date(2025, 1, 1)
        start_date = st.date_input("Start Date", value=default_start_date)
    with col2:
        default_end_date = datetime.date(2026, 2, 9)
        end_date = st.date_input("End Date", value=default_end_date)

    start_date_str = start_date.isoformat()
    end_date_str = end_date.isoformat()

    st.subheader("Average Prices within Selected Date Range")

    # Load and display average prices
    df_bitcoin = load_historical_prices(coin_id='bitcoin', start_date=start_date_str, end_date=end_date_str)
    df_oil = load_oil_prices(start_date=start_date_str, end_date=end_date_str)
    df_gspc = load_stock_prices(ticker='^GSPC', start_date=start_date_str, end_date=end_date_str)
    df_nsei = load_stock_prices(ticker='^NSEI', start_date=start_date_str, end_date=end_date_str)

    st.info(f"Bitcoin data shape: {df_bitcoin.shape}")
    st.info(f"Oil data shape: {df_oil.shape}")
    st.info(f"GSPC data shape: {df_gspc.shape}")
    st.info(f"NSEI data shape: {df_nsei.shape}")

    avg_bitcoin_price = df_bitcoin['price_inr'].mean() if not df_bitcoin.empty else 0
    avg_oil_price = df_oil['price_usd'].mean() if not df_oil.empty else 0
    avg_gspc_close = df_gspc['close'].mean() if not df_gspc.empty else 0
    avg_nsei_close = df_nsei['close'].mean() if not df_nsei.empty else 0

    st.info(f"Avg Bitcoin: {avg_bitcoin_price}")
    st.info(f"Avg Oil: {avg_oil_price}")
    st.info(f"Avg GSPC: {avg_gspc_close}")
    st.info(f"Avg NSEI: {avg_nsei_close}")

    avg_data = {
        'Asset': ['Bitcoin (INR)', 'Crude Oil (USD)', 'S&P 500 (^GSPC)', 'NIFTY 50 (^NSEI)']
    }

    if avg_bitcoin_price != 0:
        avg_data['Average Price'] = [f'{avg_bitcoin_price:,.2f}', f'{avg_oil_price:,.2f}', f'{avg_gspc_close:,.2f}', f'{avg_nsei_close:,.2f}']
    else:
        avg_data['Average Price'] = ['N/A', 'N/A', 'N/A', 'N/A']

    st.dataframe(pd.DataFrame(avg_data), hide_index=True)

    st.subheader("Daily Market Snapshot")

    # Prepare data for plotting
    # Bitcoin
    if not df_bitcoin.empty:
        df_bitcoin_plot = df_bitcoin[['date', 'price_inr']].copy()
        df_bitcoin_plot.columns = ['date', 'value']
        df_bitcoin_plot['asset'] = 'Bitcoin'

    # Crude Oil
    if not df_oil.empty:
        df_oil_plot = df_oil[['date', 'price_usd']].copy()
        df_oil_plot.columns = ['date', 'value']
        df_oil_plot['asset'] = 'Crude Oil'

    # S&P 500
    if not df_gspc.empty:
        df_gspc_plot = df_gspc[['date', 'close']].copy()
        df_gspc_plot.columns = ['date', 'value']
        df_gspc_plot['asset'] = 'S&P 500'

    # NIFTY 50
    if not df_nsei.empty:
        df_nsei_plot = df_nsei[['date', 'close']].copy()
        df_nsei_plot.columns = ['date', 'value']
        df_nsei_plot['asset'] = 'NIFTY 50'

    # Combine all plot dataframes for Altair
    all_plot_data = []
    if 'df_bitcoin_plot' in locals() and not df_bitcoin_plot.empty: all_plot_data.append(df_bitcoin_plot)
    if 'df_oil_plot' in locals() and not df_oil_plot.empty: all_plot_data.append(df_oil_plot)
    if 'df_gspc_plot' in locals() and not df_gspc_plot.empty: all_plot_data.append(df_gspc_plot)
    if 'df_nsei_plot' in locals() and not df_nsei_plot.empty: all_plot_data.append(df_nsei_plot)

    if all_plot_data:
        df_combined_plot = pd.concat(all_plot_data)

        # Create the interactive chart
        chart = alt.Chart(df_combined_plot).mark_line().encode(
            x=alt.X('date:T', axis=alt.Axis(title='Date')),
            y=alt.Y('value:Q', axis=alt.Axis(title='Price')),
            color='asset:N',
            tooltip=['date:T', 'asset:N', alt.Tooltip('value:Q', format='.2f')]
        ).properties(
            title='Daily Prices Over Time'
        ).interactive()

        st.altair_chart(chart, width='stretch')
    else:
        st.warning("No data available for the selected date range.")


# --- Page 2: SQL Query Runner ---
def sql_query_runner_page():
    st.title("SQL Query Runner")
    st.markdown("Execute predefined SQL queries against the database.")

    # Predefined queries
    predefined_queries = {
        "Top 3 Cryptocurrencies by Market Cap": "SELECT id, name, market_cap, market_cap_rank FROM cryptocurrency ORDER BY market_cap_rank ASC LIMIT 3;",
        "Coins with >90% Circulating Supply": "SELECT id, name, circulating_supply, total_supply FROM cryptocurrency WHERE (circulating_supply * 1.0 / total_supply) > 0.9 AND total_supply IS NOT NULL AND total_supply > 0;",
        "Coins within 10% of ATH": "SELECT id, name, current_price, ath FROM cryptocurrency WHERE current_price >= (ath * 0.9) AND ath IS NOT NULL;",
        "Average Market Cap Rank of High Volume Coins (>$1B)": "SELECT AVG(market_cap_rank) as average_market_cap_rank FROM cryptocurrency WHERE total_volume > 1000000000;",
        "Most Recently Updated Coin": "SELECT id, name, date FROM cryptocurrency ORDER BY date DESC LIMIT 1;",
        "Highest Bitcoin Price (Last Year)": "SELECT coin_id, date, price_inr FROM historical_prices WHERE coin_id = 'bitcoin' ORDER BY price_inr DESC LIMIT 1;",
        "Average Ethereum Price (Last Year)": "SELECT AVG(price_inr) as average_ethereum_price FROM historical_prices WHERE coin_id = 'ethereum';",
        "Daily Bitcoin Price (Feb 2025)": "SELECT date, price_inr FROM historical_prices WHERE coin_id = 'bitcoin' AND date BETWEEN '2025-02-01' AND '2025-02-28' ORDER BY date ASC;",
        "Coin with Highest Average Price (Last Year)": "SELECT coin_id, AVG(price_inr) AS average_price FROM historical_prices GROUP BY coin_id ORDER BY average_price DESC LIMIT 1;",
        "Bitcoin % Change (Sep 2025 - Jan 2026)": "SELECT ( (SELECT price_inr FROM historical_prices WHERE coin_id = 'bitcoin' AND date <= '2026-01-31' ORDER BY date DESC LIMIT 1) - (SELECT price_inr FROM historical_prices WHERE coin_id = 'bitcoin' AND date >= '2025-09-01' ORDER BY date ASC LIMIT 1) ) * 100.0 / (SELECT price_inr FROM historical_prices WHERE coin_id = 'bitcoin' AND date >= '2025-09-01' ORDER BY date ASC LIMIT 1) AS percentage_change;",
        "Highest Oil Price (Last 5 Years)": "SELECT date, price_usd FROM oil_prices WHERE date >= STRFTIME('%Y-%m-%d', DATE('now', '-5 years')) ORDER BY price_usd DESC LIMIT 1;",
        "Average Oil Price Per Year": "SELECT STRFTIME('%Y', date) AS year, AVG(price_usd) AS average_price_usd FROM oil_prices GROUP BY year ORDER BY year ASC;",
        "Oil Prices During COVID Crash (Mar-Apr 2020)": "SELECT date, price_usd FROM oil_prices WHERE date BETWEEN '2020-03-01' AND '2020-04-30' ORDER BY date ASC;",
        "Lowest Oil Price (Last 10 Years)": "SELECT date, price_usd FROM oil_prices WHERE date >= STRFTIME('%Y-%m-%d', DATE('now', '-10 years')) ORDER BY price_usd ASC LIMIT 1;",
        "Oil Price Volatility (Max-Min per Year)": "SELECT STRFTIME('%Y', date) AS year, MAX(price_usd) - MIN(price_usd) AS volatility_usd FROM oil_prices GROUP BY year ORDER BY year ASC;",
        "Highest NASDAQ Closing Price": "SELECT date, close FROM stock_prices WHERE ticker = '^IXIC' ORDER BY close DESC LIMIT 1;",
        "Top 5 S&P 500 Price Differences (High-Low)": "SELECT date, (high - low) AS price_difference FROM stock_prices WHERE ticker = '^GSPC' ORDER BY price_difference DESC LIMIT 5;",
        "Monthly Average Closing Price for Each Ticker": "SELECT ticker, STRFTIME('%Y-%m', date) AS year_month, AVG(close) AS average_closing_price FROM stock_prices GROUP BY ticker, year_month ORDER BY ticker, year_month ASC;",
        "Average NSEI Volume (2024)": "SELECT AVG(volume) AS average_volume FROM stock_prices WHERE ticker = '^NSEI' AND STRFTIME('%Y', date) = '2024';"
    }

    selected_query_name = st.selectbox("Select a predefined query:", list(predefined_queries.keys()))
    query = predefined_queries[selected_query_name]

    st.code(query, language='sql')

    if st.button("Run Query"):
        if query:
            try:
                conn = get_db_connection()
                result_df = pd.read_sql_query(query, conn)
                if not result_df.empty:
                    st.success("Query executed successfully!")
                    st.dataframe(result_df)
                else:
                    st.info("Query executed successfully, but returned no results.")
            except Exception as e:
                st.error(f"Error executing query: {e}")
        else:
            st.warning("Please select a query to run.")


# --- Page 3: Top Crypto Analysis ---
def top_crypto_analysis_page():
    st.title("Top Cryptocurrency Analysis")
    st.markdown("Visualize daily price trends and raw data for selected top cryptocurrencies.")

    df_crypto = load_cryptocurrency_data()
    top_3_coins = df_crypto.sort_values(by='market_cap_rank').head(3)
    top_3_coin_ids = top_3_coins['id'].tolist()
    top_3_coin_names = top_3_coins['name'].tolist()

    st.subheader("Select Cryptocurrencies to Analyze")
    selected_coins = st.multiselect(
        "Choose cryptocurrencies (up to 3)",
        options=df_crypto['name'].tolist(),
        default=top_3_coin_names,
        max_selections=3
    )

    if not selected_coins:
        st.warning("Please select at least one cryptocurrency.")
        return

    # Get corresponding IDs
    selected_coin_ids = df_crypto[df_crypto['name'].isin(selected_coins)]['id'].tolist()

    st.subheader("Date Range for Analysis")
    col1, col2 = st.columns(2)
    with col1:
        default_start_date = datetime.date(2025, 1, 1)
        start_date = st.date_input("Start Date", value=default_start_date, key='crypto_start_date')
    with col2:
        default_end_date = datetime.date(2026, 2, 9)
        end_date = st.date_input("End Date", value=default_end_date, key='crypto_end_date')

    start_date_str = start_date.isoformat()
    end_date_str = end_date.isoformat()

    all_historical_df = pd.DataFrame()
    for coin_id in selected_coin_ids:
        historical_df = load_historical_prices(coin_id, start_date=start_date_str, end_date=end_date_str)
        if not historical_df.empty:
            all_historical_df = pd.concat([all_historical_df, historical_df])

    if not all_historical_df.empty:
        st.subheader("Daily Price Trends")
        chart = alt.Chart(all_historical_df).mark_line().encode(
            x=alt.X('date:T', axis=alt.Axis(title='Date')),
            y=alt.Y('price_inr:Q', axis=alt.Axis(title='Price in INR')),
            color='coin_id:N',
            tooltip=['date:T', 'coin_id:N', alt.Tooltip('price_inr:Q', format='.2f')]
        ).properties(
            title='Daily Price Trend of Selected Cryptocurrencies'
        ).interactive()
        st.altair_chart(chart, width='stretch')

        st.subheader("Raw Historical Price Data")
        st.dataframe(all_historical_df)
    else:
        st.info("No historical data available for the selected cryptocurrencies in the given date range.")


# --- Main Streamlit App ---
def app():
    st.sidebar.title("Navigation")
    page = st.sidebar.radio(
        "Go to",
        ["Filters & Data Exploration", "SQL Query Runner", "Top Crypto Analysis"]
    )

    if page == "Filters & Data Exploration":
        filters_and_data_exploration_page()
    elif page == "SQL Query Runner":
        sql_query_runner_page()
    elif page == "Top Crypto Analysis":
        top_crypto_analysis_page()

if __name__ == '__main__':
    app()
```

```python
import subprocess
import time
from pyngrok import ngrok
import os
from google.colab import userdata

print("Attempting to kill any existing Streamlit processes...")
try:
    # Find process IDs listening on port 8501
    lsof_output = subprocess.check_output(['lsof', '-t', '-i:8501'], text=True)
    pids = lsof_output.strip().split('\n')
    for pid in pids:
        if pid:
            try:
                os.kill(int(pid), 9) # Kill with SIGKILL
                print(f"Killed process {pid} on port 8501.")
            except ProcessLookupError:
                print(f"Process {pid} not found.")
    time.sleep(2) # Give some time for processes to terminate
except subprocess.CalledProcessError:
    print("No processes found listening on port 8501.")
except Exception as e:
    print(f"Error while trying to kill processes: {e}")

# Start Streamlit in the background
print("Starting Streamlit app in background...")
# Redirecting output to streamlit.log
# Disabling CORS and XSRF protection for easier public access in a Colab environment
streamlit_command = [
    "streamlit", "run", "app.py",
    "--server.port", "8501",
    "--server.enableCORS", "false",
    "--server.enableXsrfProtection", "false"
]
# Use preexec_fn=os.setsid to create a new session, detaching the process from the current shell
# Redirect stdout and stderr to /dev/null to prevent hanging and clutter
process = subprocess.Popen(streamlit_command, stdout=subprocess.DEVNULL, stderr=subprocess.DEVNULL, preexec_fn=os.setsid)

time.sleep(5) # Give Streamlit some time to start

# Authenticate ngrok
print("Configuring ngrok authentication...")
NGROK_AUTH_TOKEN = userdata.get('NGROK_AUTH_TOKEN')
ngrok.set_auth_token(NGROK_AUTH_TOKEN)

# Open a ngrok tunnel to the Streamlit port
print("Opening ngrok tunnel...")
# Ensure any existing tunnels are disconnected before trying to create a new one
ngrok.kill()
tunnel = ngrok.connect(8501)
public_url = tunnel.public_url

print(f"Streamlit App is live at: {public_url}")
print("Please copy this URL and open it in your browser.")
print("If the app doesn't load correctly, try refreshing the Ngrok URL.")
```
